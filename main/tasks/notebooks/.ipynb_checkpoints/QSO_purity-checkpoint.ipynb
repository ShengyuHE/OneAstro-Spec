```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score

plt.rcParams.update({
    "font.size": 10,        # default text
    "axes.titlesize": 14,   # title
    "axes.labelsize": 12,   # axis labels
    "xtick.labelsize": 10,  # x tick labels
    "ytick.labelsize": 10,  # y tick labels
    "legend.fontsize": 10
})

result_dir = "/mnt/oss_nanhu100TB/default/zjq/results/SpecFun/result/"
```

```python
dataset = "desi-sv1" #desi-sv1
mods = ["sp", "im", "ph", "im+ph", "sp+im", "sp+ph", "sp+im+ph"][:1]
targets = ["type"]
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from sklearn.metrics import accuracy_score, confusion_matrix

cut_values = np.round(np.arange(16.5, 25.0, 0.4), 1)[10:]

metrics_dict = {
    mod: {
        "cut": [],
        "acc": [],
        "qso_recall": [],
        "n_qso": []
    }
    for mod in mods
}

for cut in cut_values:
    mask = f"R_cut_{cut:.1f}"

    for mod in mods:
        pred_file = os.path.join(
            result_dir, f"{dataset}", "class", "R_cut",
            f"{dataset}_{mod}_class_type_{mask}.npz"
        )

        if not os.path.exists(pred_file):
            print(f"Missing file: {pred_file}")
            continue

        data = np.load(pred_file, allow_pickle=True)
        y_true = data["test_true"]
        y_pred = data["test_pred"]
        classes = [str(c) for c in data["classes"]]

        acc = accuracy_score(y_true, y_pred)

        qso_recall = np.nan
        n_qso = 0

        if "QSO" in classes:
            qso_idx = classes.index("QSO")
            cm = confusion_matrix(y_true, y_pred, labels=np.arange(len(classes)))

            tp = cm[qso_idx, qso_idx]
            true_qso = cm[qso_idx, :].sum()

            qso_recall = tp / true_qso if true_qso > 0 else np.nan
            n_qso = int(true_qso)

        metrics_dict[mod]["cut"].append(cut)
        metrics_dict[mod]["acc"].append(acc)
        metrics_dict[mod]["qso_recall"].append(qso_recall)
        metrics_dict[mod]["n_qso"].append(n_qso)
```

```python
fig = plt.figure(figsize=(8, 12))
gs = gridspec.GridSpec(3, 1, figure=fig, hspace=0.08)

# ---- overall accuracy ----
ax0 = fig.add_subplot(gs[0, 0])
for mod in mods:
    ax0.plot(
        metrics_dict[mod]["cut"],
        metrics_dict[mod]["acc"],
        marker="o",
        label=mod
    )
ax0.set_title("Classification vs $R$-band magnitude cut")
ax0.set_ylabel("Overall Accuracy")
ax0.grid(alpha=0.3)
ax0.tick_params(labelbottom=False)

# ---- QSO recall ----
ax1 = fig.add_subplot(gs[1, 0], sharex=ax0)
for mod in mods:
    ax1.plot(
        metrics_dict[mod]["cut"],
        metrics_dict[mod]["qso_recall"],
        marker="o"
    )
ax1.set_ylabel("QSO Recall Accuracy")
ax1.grid(alpha=0.3)
ax1.tick_params(labelbottom=False)

# ---- number of QSOs ----
ax2 = fig.add_subplot(gs[2, 0], sharex=ax0)
for mod in mods:
    ax2.plot(
        metrics_dict[mod]["cut"],
        metrics_dict[mod]["n_qso"],
        marker="o"
    )
ax2.set_ylabel(r"$N_{\rm QSO}$")
ax2.set_xlabel(r"$R_{\text{cut}}$ magnitude (16<R<$R_{\text{cut}}$)")
ax2.grid(alpha=0.3)

handles, labels = ax0.get_legend_handles_labels()
fig.legend(
    handles, labels,
    loc="upper center",
    bbox_to_anchor=(0.5, 0.98),
    ncol=len(mods),
    frameon=False,
    fontsize=12
)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
