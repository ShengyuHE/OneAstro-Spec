```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score

plt.rcParams.update({
    "font.size": 10,        # default text
    "axes.titlesize": 12,   # title
    "axes.labelsize": 10,   # axis labels
    "xtick.labelsize": 10,  # x tick labels
    "ytick.labelsize": 10,  # y tick labels
    "legend.fontsize": 10
})

result_dir = "/mnt/oss_nanhu100TB/default/zjq/results/SpecFun/result"
```

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec

dataset = "desi-sv1"
mods = ["sp", "im", "ph", "im+ph", "sp+im", "sp+ph", "sp+im+ph"]  # 6 panels

results = {}
cms = []
accs = []
per_class_accs = []
classes_list = []


for mod in mods:
    path = os.path.join(result_dir, dataset, "class", f"{dataset}_{mod}_class_type.npz")    
    result = np.load(path, allow_pickle=True)

    y_true = result["test_true"]
    y_pred = result["test_pred"]
    classes = result["classes"]

    cm = confusion_matrix(y_true, y_pred)

    overall_acc = np.trace(cm) / np.sum(cm)

    per_class_acc = np.divide(
        cm.diagonal(),
        cm.sum(axis=1),
        out=np.zeros_like(cm.diagonal(), dtype=float),
        where=cm.sum(axis=1) != 0
    )

    cms.append(cm)
    accs.append(overall_acc)
    per_class_accs.append(per_class_acc)
    classes_list.append(classes)

# === Use same color scale ===
vmax = max(cm.max() for cm in cms)

# === Plot with GridSpec ===
fig = plt.figure(figsize=(18, 4))
gs = gridspec.GridSpec(1, len(mods), wspace=0.05)
axes = []
for i, mod in enumerate(mods):
    ax = fig.add_subplot(gs[0, i])
    axes.append(ax)

    cm = cms[i]
    classes = classes_list[i]

    im = ax.imshow(cm, cmap="Blues", vmin=0, vmax=vmax)

    # --- Title ---
    acc_text = " | ".join(
        [f"{cls}:{acc*100:.1f}" for cls, acc in zip(classes, per_class_accs[i])]
    )
    ax.set_title(f"{mod}\nAcc={accs[i]*100:.1f}\n{acc_text}", fontsize=7)

    ax.set_xticks(range(len(classes)))
    ax.set_xticklabels(classes, rotation=45, fontsize=8)

    if i == 0:
        ax.set_yticks(range(len(classes)))
        ax.set_yticklabels(classes, fontsize=8)
        ax.set_ylabel("True")
    else:
        ax.set_yticks([])

    ax.set_xlabel("Pred")

    # --- annotations ---
    threshold = cm.max() / 2
    for r in range(cm.shape[0]):
        for c in range(cm.shape[1]):
            val = cm[r, c]
            color = "white" if val > threshold else "black"
            ax.text(c, r, val, ha="center", va="center", color=color, fontsize=7)

# === Shared colorbar ===
cbar = fig.colorbar(im, ax=axes, fraction=0.02, pad=0.01, shrink=0.7)
cbar.ax.tick_params(labelsize=8)

plt.tight_layout()
plt.show()
```
