```python
def get_obj_type_index(SV1_cat,DR1_datapath,type_obj,row_max):
    if type_obj == 'BGS':
        DATA_DR1 = DR1_datapath + '/' + 'datcomb_BGS_BRIGHT_tarspecwdup_zdone.fits'
    if type_obj == 'LRG':
        DATA_DR1 = DR1_datapath + '/' + 'datcomb_LRG_tarspecwdup_zdone.fits'
    if type_obj == 'ELG':
        DATA_DR1 = DR1_datapath + '/' + 'datcomb_ELG_LOPnotqso_tarspecwdup_zdone.fits'
    if type_obj == 'QSO':
        DATA_DR1 = DR1_datapath + '/' + 'datcomb_QSO_tarspecwdup_zdone.fits'
    DR1_cat = fitsio.FITS(DATA_DR1)
    SV1_ID = SV1_cat[1].read('TARGETID')
    DR1_ID = DR1_cat[1].read('TARGETID',rows=range(0,row_max))
    _,SV1_idx,_ = np.intersect1d(SV1_ID, DR1_ID,return_indices=True)
    return SV1_idx 

def get_obj_type_col(SV1_cat,DR1_datapath):
    BG_SV1_idx = get_obj_type_index(SV1_cat,DR1_datapath,'BGS',row_max_BG)
    LRG_SV1_idx = get_obj_type_index(SV1_cat,DR1_datapath,'LRG',row_max_LRG)
    ELG_SV1_idx = get_obj_type_index(SV1_cat,DR1_datapath,'ELG',row_max_ELG)
    QSO_SV1_idx = get_obj_type_index(SV1_cat,DR1_datapath,'QSO',row_max_QSO)
    SV1_ID = SV1_cat[1].read('TARGETID')
    OBJECT_TYPE = np.zeros_like(SV1_ID)
    OBJECT_TYPE[BG_SV1_idx] = 'BGS'
    OBJECT_TYPE[LRG_SV1_idx] = 'LRG'
    OBJECT_TYPE[ELG_SV1_idx] = 'ELG'
    OBJECT_TYPE[QSO_SV1_idx] = 'QSO'
    none_obj_idx = OBJECT_TYPE == ''
    print(none_obj_idx.sum())
    OBJECT_TYPE[none_obj_idx] = 'None'
    return OBJECT_TYPE
```
